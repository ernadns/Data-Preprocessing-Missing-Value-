# removing missing value

```python
dataset1 = pd.read_csv("pima-indians-diabetes.csv")
dataset1.head()
print (dataset1.describe())
```

```python
# Melihat apakah terdapat missing value

dataset1[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']] = dataset1[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].replace(0,np.NaN)
print(dataset1.isnull().sum())
```

```python
dataset1.head(20)
```

```python
# Menghilangkan missing value
dataset1.dropna(inplace=True)
# summarize the number of rows and columns in the dataset
print dataset1.shape
```

```python
dataset2 = pd.read_csv('pima-indians-diabetes.csv')
dataset2[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']] = dataset2[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].replace(0,np.NaN)
dataset2
```

```python
#Cara mengganti data missing dengan nilai mean
dataset2['Insulin'].fillna((dataset2['Insulin'].mean()))
dataset2
```

```python
# Mengetahui banyaknya nilai NaN

dataset2.isnull().sum()
```

```python
# Mengganti data missing secara bersamaan pada 1 file dengan mean
dataset2.fillna(dataset2.mean(), inplace=True)

# Mengetahui banyaknya nilai NaN
print (dataset2.isnull().sum())
```
